```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
%matplotlib inline
```

```python
# Read csv into DataFrame
df = pd.read_csv('Sample_time_series.csv', index_col='date_time', parse_dates=True)
df.head()
```

```python
df.info()
```

```python
# Briefly plot data for visualization purposes
df.plot(y='total_support_requests')
```

```python
# Normalize data
scaler = MinMaxScaler(feature_range=(0, 1))
data = scaler.fit_transform(df.total_support_requests.values.reshape((-1, 1)))

# Create dataset with 20 features per time step
# X contains total support request info from previous 20 time steps
# y is target with total support requests of next time step
num_features = 20
X, y = [], []
for i in range(len(data) - num_features - 1):
    X.append(data[i:i+num_features, 0])
    y.append(data[i+num_features, 0])
X = np.array(X)
y = np.array(y)

# Split data into 80% training and 20% test
train_size = int(0.8 * len(data))
X_train = X[:train_size, :]
y_train = y[:train_size]
X_test = X[train_size:, :]
y_test = y[train_size:]
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# Reshape input for Keras
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))
```

```python
# Create Keras model
model = Sequential()
model.add(LSTM(100, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.5))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train, y_train, epochs=20, batch_size=50, verbose=1, shuffle=False)
model.summary()
```

```python
# Evaulate model; metric is mean squared error of normalized data
model.evaluate(X_test, y_test)
```

```python
# Predict on training and test sets
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

# Invert scaler transforms on data
train_pred = scaler.inverse_transform(train_pred)
y_train = scaler.inverse_transform(y_train.reshape((-1, 1)))
test_pred = scaler.inverse_transform(test_pred)
y_test = scaler.inverse_transform(y_test.reshape((-1, 1)))

# Compute metrics on non-normalized data
print('Training root mean squared error: ', np.sqrt(mean_squared_error(y_train, train_pred)))
print('Test root mean squared error: ', np.sqrt(mean_squared_error(y_test, test_pred)))
```

```python
# Plot actual vs. predicted data for first 500 time steps in test data
x = np.arange(500)
plt.figure(figsize=(10, 6))
plt.plot(x, y_test[:500, 0], label='actual')
plt.plot(x, test_pred[:500, 0], label='predicted')
plt.xlabel('Time Step')
plt.ylabel('Total Support Requests')
plt.legend()
```

The multivariate LSTM model seems to forecast the time series data quite well.
